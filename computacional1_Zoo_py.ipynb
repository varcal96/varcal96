## Asignamos valores a las caracteristicas.

```python
datos_zoo.loc[ datos_zoo['cabello'] <= 0, 'cabello_Encoded']= 0
datos_zoo.loc[ datos_zoo['cabello'] > 0, 'cabello_Encoded']= 1

datos_zoo.loc[ datos_zoo['pluma'] <= 0, 'pluma_Encoded']= 0
datos_zoo.loc[ datos_zoo['pluma'] > 0, 'pluma_Encoded']= 1

datos_zoo.loc[ datos_zoo['huevos'] <= 0, 'huevos_Encoded']= 0
datos_zoo.loc[ datos_zoo['huevos'] > 0, 'huevos_Encoded']= 1

datos_zoo.loc[ datos_zoo['leche'] <= 0, 'leche_Encoded']= 0
datos_zoo.loc[ datos_zoo['leche'] > 0, 'leche_Encoded']= 1

datos_zoo.loc[ datos_zoo['aerotransportado'] <= 0, 'aereo_Encoded']= 0
datos_zoo.loc[ datos_zoo['aerotransportado'] > 0, 'aereo_Encoded']= 1

datos_zoo.loc[ datos_zoo['acuatico'] <= 0, 'acuatico_Encoded']= 0
datos_zoo.loc[ datos_zoo['acuatico'] > 0, 'acuatico_Encoded']= 1

datos_zoo.loc[ datos_zoo['depredador'] <= 0, 'depredador_Encoded']= 0
datos_zoo.loc[ datos_zoo['depredador'] > 0, 'depredador_Encoded']= 1

datos_zoo.loc[ datos_zoo['dentado'] <= 0, 'dentado_Encoded']= 0
datos_zoo.loc[ datos_zoo['dentado'] > 0, 'dentado_Encoded']= 1

datos_zoo.loc[ datos_zoo['columna vert'] <= 0, 'columnaVert_Encoded']= 0
datos_zoo.loc[ datos_zoo['columna vert'] > 0, 'columnaVert_Encoded']= 1

datos_zoo.loc[ datos_zoo['respira'] <= 0, 'respira_Encoded']= 0
datos_zoo.loc[ datos_zoo['respira'] > 0, 'respira_Encoded']= 1

datos_zoo.loc[ datos_zoo['venenoso'] <= 0, 'venenoso_Encoded']= 0
datos_zoo.loc[ datos_zoo['venenoso'] > 0, 'venenoso_Encoded']= 1

datos_zoo.loc[ datos_zoo['aletas'] <= 0, 'aletas_Encoded']= 0
datos_zoo.loc[ datos_zoo['aletas'] > 0, 'aletas_Encoded']= 1

datos_zoo.loc[ datos_zoo['patas'] <= 0, 'patas_Encoded']= 0
datos_zoo.loc[(datos_zoo['patas'] > 0) & (datos_zoo['patas'] < 3), 'patas_Encoded'] = 2
datos_zoo.loc[(datos_zoo['patas'] > 3) & (datos_zoo['patas'] < 5), 'patas_Encoded'] = 4
datos_zoo.loc[(datos_zoo['patas'] > 4) & (datos_zoo['patas'] < 6), 'patas_Encoded'] = 5
datos_zoo.loc[(datos_zoo['patas'] > 5) & (datos_zoo['patas'] < 7), 'patas_Encoded'] = 6
datos_zoo.loc[ datos_zoo['patas'] > 7, 'patas_Encoded']= 8

datos_zoo.loc[ datos_zoo['cola'] <= 0, 'cola_Encoded']= 0
datos_zoo.loc[ datos_zoo['cola'] > 0, 'cola_Encoded']= 1

datos_zoo.loc[ datos_zoo['domestico'] <= 0, 'domestico_Encoded']= 0
datos_zoo.loc[ datos_zoo['domestico'] > 0, 'domestico_Encoded']= 1

datos_zoo.loc[ datos_zoo['tamaño de gato'] <= 0, 'tamGato_Encoded']= 0
datos_zoo.loc[ datos_zoo['tamaño de gato'] > 0, 'tamGato_Encoded']= 1
```

##Podado del árbol (pruning)

Aunque inicialmente se ha empleado un valor de "max_depth=5", este no tiene por qué ser el mejor valor. Con el objetivo de identificar la profundidad óptima que consigue reducir la varianza y aumentar la capacidad predictiva del modelo, se somete al árbol a un proceso de pruning

```python
# Post pruning (poda de complejidad constante) por validación cruzada
# ------------------------------------------------------------------------------
# Valores de ccp_alpha evaluados
param_grid = {'ccp_alpha':np.linspace(0, 5, 10)}

# Búsqueda por validación cruzada
grid = GridSearchCV(
        # El árbol se crece al máximo posible antes de aplicar el pruning
        estimator = DecisionTreeClassifier(
                            max_depth         = None,
                            min_samples_split = 2,
                            min_samples_leaf  = 1,
                            random_state      = 100
                       ),
        param_grid = param_grid,
        scoring    = 'accuracy',
        cv         = 10,
        refit      = True,
        return_train_score = True
      )

grid.fit(X_train, y_train)

fig, ax = plt.subplots(figsize=(6, 3.84))
scores = pd.DataFrame(grid.cv_results_)
scores.plot(x='param_ccp_alpha', y='mean_train_score', yerr='std_train_score', ax=ax)
scores.plot(x='param_ccp_alpha', y='mean_test_score', yerr='std_test_score', ax=ax)
ax.set_title("Error de validacion cruzada vs hiperparámetro ccp_alpha");
```

```python
# Mejor valor ccp_alpha encontrado
# ------------------------------------------------------------------------------
grid.best_params_
```

```python
# Estructura del árbol final
# ------------------------------------------------------------------------------
modelo_final = grid.best_estimator_
print(f"Profundidad del árbol: {modelo_final.get_depth()}")
print(f"Número de nodos terminales: {modelo_final.get_n_leaves()}")

fig, ax = plt.subplots(figsize=(14, 10))
plot = plot_tree(
            decision_tree = clasificador,
            feature_names = list(zoo_encoded.drop(['tipo'], axis=1)),
            class_names   = 'tipo',
            filled        = True,
            impurity      = False,
            ax            = ax
       )
```
